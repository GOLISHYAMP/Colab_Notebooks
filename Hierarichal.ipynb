```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
dataset = load_iris()
```

```python
dataset.keys()
```

```python
df = pd.DataFrame(data = dataset.data, columns=dataset.feature_names)
```

```python
df.head()
```

Lets apply PCA on the above features so that we can visualized the data in 2D, lets derive two features

```python
from os import XATTR_CREATE
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
df = pca.fit_transform(df)
```

```python
df
```

Lets visualize the data

```python
plt.scatter(df[:, 0], df[:, -1])
```

By looking at the dataset i could able to see clearly there are two clusters, but we know in the iris dataset we have 3 types of flowers. so lets see.

```python
import scipy.cluster.hierarchy as sc
plt.figure(figsize=(20,7))
plt.title('Dendrogram')
sc.dendrogram(sc.linkage(df, method = 'ward'))
plt.xlabel('Sample Index')
plt.ylabel('Eucledian Distance')
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
cluster.fit(df)
```

```python
cluster.labels_
```

```python
plt.scatter(df[:, 0], df[:, -1], c = cluster.labels_)
```

So, we have saw how we have found the labels for the dataset using the agglomerative hirarichal clustering.

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
cluster.fit(df)
```

```python
cluster.labels_
```

```python
plt.scatter(df[:, 0], df[:, -1], c = cluster.labels_)
```
