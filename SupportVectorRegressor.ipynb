```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = sns.load_dataset('tips')
```

# **So, here my dependent variable will be the total_bill and rest will be the independent variables.**
I have to make the categorical data into the numeric form.\
For that i have to do the label encoder and one hot encoder

```python
dataset['sex'].value_counts()
```

```python
dataset['smoker'].value_counts()
```

```python
dataset['day'].value_counts()
```

```python
dataset['time'].value_counts()
```

So, I can use the label encoder on sex, smoker and time\
and one hot encoder on day

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dataset['sex'] = le.fit_transform(dataset['sex'])
dataset['smoker'] = le.fit_transform(dataset['smoker'])
dataset['time'] = le.fit_transform(dataset['time'])
```

```python
dataset
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('hot_encoder', OneHotEncoder(drop='first'), [4])], remainder='passthrough')
```

```python
dataset = ct.fit_transform(dataset)
```

```python
dataset[:5, :]
```

```python
df = pd.DataFrame(dataset)
```

```python
df.head()
```

```python
X = df[[0,1,2,4,5,6,7,8]]
```

```python
X
```

```python
y = df.iloc[:, 3]
```

```python
print(type(y))
y
```

# **Train test split of the dataset**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state = 31)
```

# **Model training**

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(X_train, y_train)
y_pred = svr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print(mean_squared_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```

# **Very less accuracy, lets do hyperparameter tunning using gridsearchCV**

```python
params = {
    'kernel' : ['linear', 'rbf', 'poly', 'sigmoid'],
    'C' : [100, 10, 1, 0.1, 0.01, 0.001],
    'gamma' : [1, 0.1, 0.01, 0.001],
    'degree' : [2, 3]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator=SVR(), param_grid=params, scoring='r2',n_jobs=-1, refit = True, cv = 5)
grid.fit(X_train, y_train)
```

```python
y_pred = grid.predict(X_test)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
from sklearn.metrics import r2_score, mean_squared_error
print(r2_score(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
```
